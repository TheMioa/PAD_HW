```python
import pandas as pd

```

Zad 1

```python
df = pd.read_csv('PAD_03_PD.csv', sep = ';')
df.groupby('Country').count()
```

```python
#To delete NaNs from the df:
df.dropna(inplace=True)
df.groupby('Country').count()
```

Zad 2

```python
df['owned_goods'] = df.owns_TV + df.owns_Phone + df.owns_car + df.owns_house
df
```

Zad 3

```python
df.groupby('gender').mean().owned_goods.round(2)
```

Zad 4

```python
meanOwnedByCountry = df.groupby('Country').mean().owned_goods.reset_index()
minAge = df.groupby('Country').min().Age.reset_index()
result = pd.merge(meanOwnedByCountry, minAge, how='outer', on='Country')
result
```
